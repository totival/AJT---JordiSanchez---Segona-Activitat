```python
llista_radis=list(range(100))
print(llista_radis)
```

```python
import math
llista_perimetres=[2*math.pi*radi for radi in llista_radis]
print(llista_perimetres)
```

```python
import math
llista_areas=[math.pi*radi**2 for radi in llista_radis]
print(llista_areas)
```

```python
import math
llista_volums=[(4/3)*math.pi*radi**3 for radi in llista_radis]
print(llista_volums)
```

```python
perimetres_parells = [perimetre for radi, perimetre in zip(llista_radis, llista_perimetres) if radi % 2 == 0]
print("Perímetres amb radis parells:")
print(perimetres_parells)

areas_multiples_3 = [area for radi, area in zip(llista_radis, llista_areas) if radi % 3 == 0]
print("\nÀrees amb radis múltiples de 3:")
print(areas_multiples_3)

ultims_5_volums = llista_volums[-5:]
print("\nVolums dels últims 5 radis:")
print(ultims_5_volums)
```

```python
noms=["  anna  ", "MARC", "joan  ",  "  Laia" ]
llista_noms_nets=[nom.strip() for nom in noms]
print(llista_noms_nets)
```

```python
llista_noms_nets_2=[nom.capitalize() for nom in llista_noms_nets]
print(llista_noms_nets_2)
```

```python
llargades_originals=[len(nom) for nom in noms]
print(llargades_originals)
```

```python
llargades_finals=[len(nom.strip().capitalize()) for nom in noms]
print(llargades_finals)
```

```python
anna_orig=len([nom for nom in noms if "anna" in nom.lower()][0])
joan_orig=len([nom for nom in noms if "joan" in nom.lower()][0])

anna_final=len([nom for nom in llista_noms_nets_2 if "anna" in nom.lower()][0]) 
joan_final=len([nom for nom in llista_noms_nets_2 if "joan" in nom.lower()][0])

print(f"Diferència original entre anna i joan: {anna_orig - joan_orig}")
print(f"Diferència final entre Anna i Joan: {anna_final - joan_final}")
```
